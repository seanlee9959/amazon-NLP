```python
import requests
from bs4 import BeautifulSoup
import scrapy

# Define the URL for the product page you want to scrape
productID = 'B0B9HNDVXJ'
url = 'https://www.amazon.com/dp/'+productID +'/#customerReviews'
#url = 'https://www.amazon.in/Signature-Godzilla-Citrulline-Beta-Alanine-Strawberry/dp/B0B9HNDVXJ/ref=sr_1_6?keywords=godzilla%2Bpre%2Bworkout&qid=1677804489&sprefix=godzilla%2Bpre%2Caps%2C137&sr=8-6&th=1'

# Send a GET request to the URL and extract the HTML content
response = requests.get(url, headers = HEADERS)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.prettify
```

```python
# Find all the review containers on the page
review_containers = soup.find_all('div', {'class': 'a-section review-section'})

# Iterate over each review container and extract the relevant information
for review in review_containers:
    # Extract the reviewer's name
    reviewer_name = review.find('span', {'class': 'a-profile-name'}).get_text()
    
    # Extract the date of the review
    review_date = review.find('span', {'class': 'review-date'}).get_text()
    
    # Extract the star rating
    star_rating = review.find('span', {'class': 'a-icon-alt'}).get_text()
    
    # Extract the text of the review
    review_text = review.find('span', {'class': 'review-text'}).get_text()
    
    # Print out the extracted information for each review
    print(f'Reviewer Name: {reviewer_name}')
    print(f'Review Date: {review_date}')
    print(f'Star Rating: {star_rating}')
    print(f'Review Text: {review_text}\n')
```
